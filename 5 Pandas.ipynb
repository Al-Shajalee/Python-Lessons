####  <font color=orange>IQR method</font>


https://towardsdatascience.com/understanding-boxplots-5e2df7bcbd51

```python
# outliers_detecter
def outliers_IQR(df, col):
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR= Q3-Q1
    
    lower_bound= Q1-1.5*IQR
    upper_bound= Q3+1.5*IQR
    
    outliers_index=list( df.index[ (df[col] < lower_bound) | (df[col] > upper_bound) ] )
    
    return outliers_index
```

```python
# A dictionary of outliers_columns_index
col_list= ["Salary", "V"]
outliers_index={col:outliers_IQR(df, col) for col in col_list}
outliers_index
```

```python
# returning a list of outliers without repeatetion
col_list= ["Salary", "V"]
outliers_index_list = list(set([ item for elem in outliers_index.values() for item in elem]))
outliers_index_list
```

###  <font color=green>Remove Outliers</font>

```python
def outliers_remove(df, outliers_index):
    # dropping the outliers from the data fram
    df_wo_outliers=df.drop(outliers_index_list)
    
    return df_wo_outliers
```

```python
df_wo_outliers=outliers_remove(df, outliers_index)
df_wo_outliers
```

```python
# Reset the index of "df" 
df_wo_outliers = df_wo_outliers.reset_index(drop=True, inplace=True) 
df_wo_outliers
```

```python
# outliers_detecter
def outliers_outliers_remove(df, col_list):
    
    # outliers_detecter: IQR method
    def outliers_IQR(df, col):
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR= Q3-Q1
    
        lower_bound= Q1-1.5*IQR
        upper_bound= Q3+1.5*IQR
    
        outliers_index=list( df.index[ (df[col] < lower_bound) | (df[col] > upper_bound) ] )
    
        return outliers_index
    
    
    # A dictionary of outliers_index
    outliers_index_dic={col:outliers_IQR(df, col) for col in col_list}
    #print("outliers_index_dic=" , outliers_index_dic)
    
    # A list of outliers_index without repeatetion
    outliers_index_list = list(set([ item for elem in outliers_index_dic.values() for item in elem]))
    #print("outliers_index_list = " , outliers_index_list)
    
    # Dropping the outliers from the data fram
    df_wo_outliers=df.drop(outliers_index_list)
    
    
    # Reset the index of the data fram
    df_wo_outliers = df_wo_outliers.reset_index(drop=True) 

    return df_wo_outliers
```

```python
col_list= ["Salary", "V"]
df_wo_outliers = outliers_outliers_remove(df, col_list)
df_wo_outliers
```

    Combining Data Across Rows or Columns

    Keep in mind that dimensions should match along the axis you are concatenating on. 

###  <font color=green>Vertically</font>

    Compile data vertically on each other (for data frames having similar column names).


```python
pd.concat([df_a, df_b, df_c])
```

###  <font color=green>Horizontally</font>

    Note: choosing the right axes to concatenate data frames is necessary.

##  <font color=blue>Merging</font>

    The merge function allows you to merge DataFrames together using a similar logic as merging SQL Tables together.


```python
left=df[["Company", "Name"]].iloc[0:3]
left["key"]=['k0', 'k1', 'k2']
left
```

```python
right=df[["Age", "Salary"]].iloc[0:3]
right["key"]=['k0', 'k1', 'k3']
right
```

###  <font color=green>"One key column: "key" column is similar</font>

Note: Key column should have only uniques values

```python
pd.merge(left, right)
```

```python
pd.merge(left, right, how="inner", on="key")
```

###  <font color=green>Multi key columns</font>

```python
left=df[["Company", "Name"]].iloc[0:3]
left["key1"]=['k0', 'k0', 'k1']
left["key2"]=['k0', 'k1', 'k0']
left
```

```python
right=df[["Age", "Salary"]].iloc[0:3]
right["key1"]=['k0', 'k1', 'k2']
right["key2"]=['k0', 'k0', 'k0']
right
```

####  <font color=orange> Inner</font>

    Merging only rows having similar key1 and key 2
    Note: Columns are diffrent 


```python
pd.merge(left, right)
```

```python
pd.merge(left, right, how="inner", on=["key1", "key2"])
```

####  <font color=orange> Outer</font>

Merging similar and non-similar keys

```python
pd.merge(left, right, how="outer", on=["key1", "key2"])
```

```python
pd.merge(left, right, how="left", on=["key1", "key2"])
```

```python
pd.merge(left, right, how="right", on=["key1", "key2"])
```

##  <font color=blue>Joining</font>

Joining is a convenient method for combining the columns of two potentially differently-indexed DataFrames into a single result DataFrame.


```python
left=df[["Company", "Name"]].iloc[0:3]
left.index=['k0', 'k1', 'k2']
left
```

```python
right=df[["Age", "Salary"]].iloc[1:4]
right.index=['k1', 'k2', 'k3']
right
```

###  <font color=green>Inner</font>

```python
left.join(right)
```

###  <font color=green>Outer</font>

```python
left.join(right, how="outer")
```
